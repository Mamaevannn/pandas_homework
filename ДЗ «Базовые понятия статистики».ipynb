Проверив полученные значения на соответствие значениям из описания данных, можно увидеть несоответствие в данных столбца ***age***  
Согласно описанию данных, в столбце age могут встречаться только значения 1(adult) или 2(young), а в представленном датафрейме представлены значения 1 и 9.  
Рассчитав средние значения показателей для лошадей с возрастом 1 и 9, можно заметить, что значительно различаются средние значения для пармаетра pulse. В описании указано, что для взрослых лошадей норма пульса колеблется от 30 до 40, а для молодых, как правило, характерны более высокие значения.  
Соответсвенно, можно предположить, что это ошибка в данных, и целесообразно заменить все значения 9 на 2. 

```python
horses[horses.age == 1].mean()
```

```python
horses[horses.age == 9].mean()
```

```python
horses.loc[(horses.age == 9), 'age'] = 2 #заменяем 9 на 2
```

```python
#выбросы
def find_outlier(df, param):
    q1 = df[param].quantile(0.25)
    q3 = df[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    df2= df[~df[param].between(lower_bound, upper_bound, inclusive=True)]
    return df2.loc[df2[param].notnull()]
```

```python
find_outlier(horses, 'pulse') #выбросы для столбца pulse

# в выбросы попали очень высокие значения пульса, которые могут быть вызваны болезненными поражениями или циркуляторным шоком,
# что вполне вероятно, учитывая преимущественно высокие значения respiratory_rate, rectaltemperature, pain 
# также обращает на себя внимание то, что в выборку попали только взрослые лошадии 
# и тот факт, что все из них, кроме 41 (которой единственной из выборки не делали операцию) в итоге умерли
# также стоит обратить внимание, что в выборку попали только взрослые лошади
```

```python
find_outlier(horses,'rectal_temperature').sort_values(by=['rectal_temperature']) #выбросы для столбца rectaltemperature

# норма данного показателя - 37.8. в выбросы попали как высокие, так и низкие значения
# высокие значения, согласно описанию данных, могут быть связаны с инфекцией, а низкие - с последними стадиями шоками 
# примечательно, что в выборку попали преимущественно молодые лошади 
```

```python
find_outlier(horses,'respiratory_rate').sort_values(by=['respiratory_rate']) #выбросы для столбца respiratory_rate

# в выборку попали лошади с очень высокими значениями частоты дыхания, которые превышают норму в 6-10 раз
# вероятно, у этих лошадей были диагнастированы проблемы с дыхаиельными путями,которые вызывают такие отклонения
```

```python
# находим пустые значение в столбце surgery?
horses[horses['surgery?'].isna()]
```

Беря в расчет наблюдаемую тенденцию - чем выше показатель боли, тем больше вероятность того, что потребовалось хирургическое вмешательство (медиана в pain выше у тех, кому проводилась операция) - можно предположить, что операция состоялась, поскольку в данном случае значение равно 4 (периодические мучительные боли)

```python
horses_filled = horses.copy()
horses_filled['surgery?'] = horses_filled['surgery?'].fillna(1) #заполняем пропуск
```

**outcome**

```python
horses_filled[horses['outcome'].isna()]
```

```python
horses_filled.outcome.value_counts()
```

учитывая, что все показатели не сильно отклоняются от нормы и тот факт, что 60% всех лошадей в итоге выжило, можем предположить (и надеяться), что и в этом случае все обошлось благополучно

```python
horses_filled['outcome'] = horses_filled['outcome'].fillna(1) #заполняем пропуск
```

**rectal_temperature**

```python
horses[horses.rectal_temperature.isna()]
```

Поскольку существует корреляция (согласно описанию данных) между значениями temperature_of_extremities и rectal_temperature, предлагается заполнить пропуски медианой в разрезе значений temperature_of_extremities

```python
horses_filled['rectal_temperature'] = horses_filled['rectal_temperature'].fillna(
    horses.groupby('temperature_of_extremities')['rectal_temperature'].transform('median'))
```

однако в столбце temperature_of_extremities также много пропущенных данных, соответственно, не получится заполнить все пропуски только по этому значению
для оставшихся пропусков можем использовать медиану в разрезе outcome

```python
horses_filled['rectal_temperature'] = horses_filled['rectal_temperature'].fillna(
    horses.groupby('outcome')['rectal_temperature'].transform('median'))
```

**pulse, respiratory_rate**

как видно из представленных выше данных, значения pulse и respiratory_rate значительно различаются в разрезе age и surgery?
следовательно, можем заполнить пропуски по такой логике

```python
horses_filled['pulse'] = horses_filled['pulse'].fillna(
    horses.groupby(['age', 'outcome'])['pulse'].transform('median'))
```
